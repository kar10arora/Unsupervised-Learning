# Data Partitioning

```python
X = df_scaled
y = df['status_type']
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss = [] # With Cluster Sum Squares
for i in range(1 , 11):
    kmeans = KMeans(
        n_clusters = i, # No of clusters
        init='k-means++',# Random Initialization Type
        n_init = 10,
        max_iter=300 , 
        random_state = 42
    )
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # inertia_ => with cluster sum squares
```

```python
plt.figure(figsize = (4,4))
plt.plot(np.arange(1,11) , wcss )
plt.title("With_Cluster_Sum_Squares")
plt.xlabel("N_Clusters")
plt.ylabel("WCSS")
plt.show()
```

# Optimal No of Clusters - 4

```python
kmeans_ = KMeans(
        n_clusters = 4, # No of clusters
        init='k-means++',# Random Initialization Type
        n_init=10,
        max_iter=300 , 
        random_state = 42
    )
y_means = kmeans_.fit(X)
y_means
```

```python
ymeans = kmeans_.fit_predict(X)
```

```python
labels = kmeans_.labels_
correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
print('Accuracy score: ',(correct_labels/float(y.size))*100,'%')
```
